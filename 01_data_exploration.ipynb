```python

import os
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import requests
import zipfile
import shutil


# ‰∏ãËΩΩÂú∞ÂùÄ
url = "http://download.tensorflow.org/data/speech_commands_v0.01.tar.gz"
data_dir = "../data"
archive_path = os.path.join(data_dir, "speech_commands_v0.01.tar.gz")

# Â¶ÇÊûúÊ≤°‰∏ãËΩΩËøáÂ∞±‰∏ãËΩΩ
if not os.path.exists(archive_path):
    print("‚è¨ ‰∏ãËΩΩËØ≠Èü≥Êï∞ÊçÆÈõÜ‰∏≠...")
    r = requests.get(url, stream=True)
    with open(archive_path, 'wb') as f:
        shutil.copyfileobj(r.raw, f)
    print("‚úÖ ‰∏ãËΩΩÂÆåÊàêÔºÅ")

# Ëß£Âéã
extract_path = os.path.join(data_dir, "speech_commands")
if not os.path.exists(extract_path):
    print("üì¶ Ê≠£Âú®Ëß£ÂéãÊï∞ÊçÆ...")
    os.makedirs(extract_path, exist_ok=True)
    import tarfile
    with tarfile.open(archive_path, 'r:gz') as tar:
        tar.extractall(path=extract_path)
    print("‚úÖ Ëß£ÂéãÂÆåÊàêÔºÅ")

# ÂàóÂá∫ÂâçÂá†‰∏™ÂëΩ‰ª§Á±ªÂà´ÔºàÊØè‰∏™ÊòØ‰∏Ä‰∏™ÁõÆÂΩïÔºâ
commands = [d for d in os.listdir(extract_path) if os.path.isdir(os.path.join(extract_path, d))]
print("ÂèØËØÜÂà´ÁöÑÂÖ≥ÈîÆËØçÁ±ªÂà´ÔºàÈÉ®ÂàÜÔºâ:")
print(commands[:10])


import IPython.display as ipd

# Èöè‰æøÈÄâ‰∏Ä‰∏™ÂÖ≥ÈîÆËØçÔºåÊØîÂ¶Ç "yes"
sample_path = os.path.join(extract_path, "yes", "0a7c2a8d_nohash_0.wav")

# Êí≠ÊîæËØ≠Èü≥
ipd.Audio(sample_path)

# ÂèØËßÜÂåñÊ≥¢ÂΩ¢
y, sr = librosa.load(sample_path)
plt.figure(figsize=(10, 3))
librosa.display.waveshow(y, sr=sr)
plt.title("Waveform of 'yes'")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()



mel_spec = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
mel_db = librosa.power_to_db(mel_spec, ref=np.max)

plt.figure(figsize=(10, 4))
librosa.display.specshow(mel_db, sr=sr, x_axis='time', y_axis='mel')
plt.colorbar(format="%+2.0f dB")
plt.title("Mel Spectrogram of 'yes'")
plt.tight_layout()
plt.show()
```

```python
import os
import librosa
import numpy as np
from sklearn.preprocessing import LabelEncoder

# ÊåáÂÆöÊï∞ÊçÆË∑ØÂæÑÂíåÁ±ªÂà´
extract_path = "../data/speech_commands"
class_names = ["yes", "no", "up", "down", "left", "right", "on", "off", "stop", "go"]
X, y = [], []

for label in class_names:
    class_dir = os.path.join(extract_path, label)
    if not os.path.isdir(class_dir):
        continue
    for file in os.listdir(class_dir)[:100]:  # ÊØèÁ±ªÊúÄÂ§öÂ§ÑÁêÜ100‰∏™Êñá‰ª∂
        if file.endswith(".wav"):
            file_path = os.path.join(class_dir, file)
            try:
                y_audio, sr = librosa.load(file_path, sr=None)
                mel_spec = librosa.feature.melspectrogram(y=y_audio, sr=sr, n_mels=128, fmax=8000)
                mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)
                X.append(mel_spec_db)
                y.append(label)
            except Exception as e:
                print(f"Â§ÑÁêÜ {file_path} Âá∫ÈîôÔºö{e}")

print(f"ÊàêÂäüÊèêÂèñÊ†∑Êú¨Êï∞Ôºö{len(X)}ÔºåÁ¨¨‰∏Ä‰∏™Ê†∑Êú¨ shapeÔºö{X[0].shape}")

# Áªü‰∏ÄÊó∂Èó¥Áª¥Â∫¶ÔºàÁ¨¨‰∫åÁª¥ÔºâÈïøÂ∫¶Ôºå‰ΩøÁî® numpy ÊâãÂä®padding
max_time_steps = max([x.shape[1] for x in X])
n_mels = X[0].shape[0]

X_padded = np.zeros((len(X), n_mels, max_time_steps), dtype=np.float32)

for i, x in enumerate(X):
    length = x.shape[1]
    X_padded[i, :, :length] = x  # ÂêéÈù¢Ë°•0

# ËΩ¨Êç¢ÂΩ¢Áä∂‰ª•ÈÄÇÈÖç Conv2D ËæìÂÖ• (samples, height, width, channels)
X_padded = X_padded[..., np.newaxis]

# Ê†áÁ≠æÁºñÁ†Å‰∏∫Êï¥Êï∞ÔºåÂÜçËΩ¨Êç¢‰∏∫ one-hot
le = LabelEncoder()
y_int = le.fit_transform(y)

from tensorflow.keras.utils import to_categorical
y_cat = to_categorical(y_int, num_classes=len(class_names))

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_padded, y_cat, test_size=0.2, random_state=42, stratify=y_int
)
print(f"ËÆ≠ÁªÉÈõÜ shape: {X_train.shape}, ÊµãËØïÈõÜ shape: {X_test.shape}")

import tensorflow as tf
from tensorflow.keras import layers, models

model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=X_train.shape[1:]),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(len(class_names), activation='softmax')
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()

history = model.fit(
    X_train, y_train,
    epochs=10,
    batch_size=32,
    validation_data=(X_test, y_test)
)
                
```
